```python
import numpy as np
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import cv2
```

```python
# Load the .h5 model
cnn = tf.keras.models.load_model('model_inception.h5')

```

```python
# Test Image Visualization
image_path = '3.jpg'
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Converting BGR to RGB

# Display the image
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# Load and resize the image to match the model's expected input size
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(224, 224))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Convert single image to a batch
```

```python
# Make predictions
predictions = cnn.predict(input_arr)
```

```python
print(predictions)
```

```python
def get_prediction_label(predictions, class_names):
    """
    This function returns the class label corresponding to the highest prediction value.

    Parameters:
    predictions (numpy array): The output of the model's prediction.
    class_names (list): List of class labels corresponding to the model's output classes.

    Returns:
    str: The predicted class label in a readable format.
    """
    # Get the index of the highest predicted probability
    result_index = np.argmax(predictions)
    
    # Map the index to the class name
    predicted_label = class_names[result_index]
    
    # Return the readable format
    return f"Predicted Disease: {predicted_label}"

# Example usage:
class_names = [
    'Tomato___Tomato_mosaic_virus',
    'Tomato___Early_blight',
    'Tomato___Septoria_leaf_spot',
    'Tomato___Bacterial_spot',
    'Tomato___Target_Spot',
    'Tomato___Spider_mites Two-spotted_spider_mite',
    'Tomato___Tomato_Yellow_Leaf_Curl_Virus',
    'Tomato___Late_blight',
    'Tomato___healthy',
    'Tomato___Leaf_Mold'
]
```

```python
get_prediction_label(predictions, class_names)
```
